```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import plotly.express as px
```

```python
df_conversion_rate = pd.read_csv('/content/drive/MyDrive/Prueba técnica/Rocketla/Base Conversión.csv')
df_information_flux = pd.read_csv('/content/drive/MyDrive/Prueba técnica/Rocketla/InformationFlux.csv')
df_source = pd.read_csv('/content/drive/MyDrive/Prueba técnica/Rocketla/funnel_source.csv')
```

Primer Funnel Usuarios para poder entender el Flujo y nuestra tasa de Retención desde que entran hasta que convierten

```python
fig1= px.funnel(y = ['Enter the Site', 'Start Process', 'Finish Process'],
               x= [df_conversion_rate.amplitude_id.count(),df_conversion_rate.user_id.count(), 
                        df_conversion_rate['gp:Nip Verify'].count()], color_discrete_sequence =['#7a4bfb'])
fig2= px.funnel(y = ['Enter the Site', 'Start Process', 'Finish Process'],
               x= [(df_conversion_rate.amplitude_id.count()/df_conversion_rate.amplitude_id.count()*100),
                   round((df_conversion_rate.user_id.count()/df_conversion_rate.amplitude_id.count()*100),2), 
                   round((df_conversion_rate['gp:Nip Verify'].count()/df_conversion_rate.amplitude_id.count()*100),2)],
          color_discrete_sequence =['#7a4bfb'])
fig3 = px.funnel(df_source, x ='Count', y = 'Stage', color = 'Source')
fig4 = px.funnel(df_information_flux, x='Users', y = 'Process', color= 'Preventa', 
          color_discrete_sequence=['#7a4bfb','#A34BFA'])
```

```python
fig1.show()
fig2.show()
fig3.show()
fig4.show()
```
